```python
import pandas as pd
import os
```

```python
csv_path = os.path.join('..', 'database', 'Absenteeism_data.csv')
```

```python
raw_dataFrame = pd.read_csv(csv_path)
df = raw_dataFrame.copy()
df = df.drop(['ID'], axis= 1)
```

```python
sorted(df['Reason for Absence'].unique())
```

```python
reason_columns = pd.get_dummies(df['Reason for Absence'])
reason_columns['check'] = reason_columns.sum(axis=1)
reason_columns['check'].sum(axis=0)
```

```python
reason_columns['check'].unique()
```

```python
reason_columns = reason_columns.drop(['check'], axis = 1)
reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first = True)
```

```python
df = df.drop(['Reason for Absence'], axis = 1)
```

```python
reason_type_1 = reason_columns.loc[:, 1:14].max(axis=1)
reason_type_2 = reason_columns.loc[:, 15:17].max(axis=1)
reason_type_3 = reason_columns.loc[:, 18:21].max(axis=1)
reason_type_4 = reason_columns.loc[:, 22:].max(axis=1)
```

```python
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis=1)
df
```

```python
df.columns.values
columns_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
df.columns = columns_names
```

```python
columns_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
df = df[columns_names_reordered]
df.head()
```

```python
df_reason_mod = df.copy()  # new checkpoint
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format = '%d/%m/%Y')
```

```python
list_months = []
for i in range(df_reason_mod.shape[0]):
    list_months.append(df_reason_mod['Date'][i].month)
df_reason_mod['Month Value'] = list_months
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_reason_mod['Day of the Week'] = df_reason_mod['Date'].apply(date_to_weekday)
```

```python
df_reason_date_mod = df_reason_mod.drop(['Date'], axis=1)
cols = df_reason_date_mod.columns.tolist()
# cols
cols = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month Value', 'Day of the Week', 'Transportation Expense', 'Distance to Work',
 'Age', 'Daily Work Load Average', 'Body Mass Index', 'Education', 'Children', 'Pets', 'Absenteeism Time in Hours']
df_reason_date_mod = df_reason_date_mod[cols]
df_reason_date_mod.head(10)
```

```python
df_reason_date_mod['Education'].value_counts()
df_reason_date_mod['Education'] = df_reason_date_mod['Education'].map({1:0, 2:1, 3:1, 4:1})
df_reason_date_mod['Education'].unique()
```

## Final Chackpoint

```python
df_preprocessed = df_reason_date_mod.copy()
```

```python
df_preprocessed.to_csv('../database/Absenteeism_preprocessed.csv', index=False)
```
